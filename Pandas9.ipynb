**"Values_count()":-**

```python
import pandas as pd
```

```python
data = pd.read_csv("Toyota.csv")
data['Price'].value_counts()
```

```python
print(data['FuelType'].value_counts())
print("\t")
print(data['FuelType'].value_counts(normalize=True))  # Gives output in 'percentage'
```

```python
print(data['FuelType'].value_counts(normalize=True).plot())
```

```python
print(data['FuelType'].value_counts(normalize=True).plot(kind='bar'))   # Vertical Bar
```

```python
print(data['FuelType'].value_counts(normalize=True).plot(kind='barh'))  # Horizontal Bar
```

```python
print(data['FuelType'].value_counts(ascending=True).plot(kind='bar')) # Output in Ascending order
```

```python
data[['FuelType','Age']].value_counts()   # Value_counts() between two columns
```

```python
data1 = data[data['FuelType'] =='Diesel']
data1['CC'].value_counts(normalize=True).plot(kind='pie')
```

```python
data1 = data[data['FuelType'] =='Petrol']
data1['CC'].value_counts(normalize=True).plot(kind='pie')
```

```python
data['Age'].value_counts(bins=5)
```

```python
data['Age'].value_counts(bins=5).plot(kind='bar')
```

```python
data['Age'].value_counts(bins=5).plot(kind='pie')
```
