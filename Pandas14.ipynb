```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = pd.read_csv('Toyota.csv')
```

**"Scatter Plot":-**

```python
plt.scatter(a['Age'], a['Price'] , c= 'red')
plt.title("Scatter plot of Price vs Age")
plt.xlabel('Age(months)')
plt.ylabel('Price(Euros)')
plt.show()
```

**"Histogram":-**

```python
plt.hist(a['KM'], color= 'green',edgecolor='white',bins=6)
plt.title("Histogram of Kilometer")
plt.xlabel("Kilometer")
plt.ylabel("Frequency")
plt.show()
```

**"Bar Plots":-**

```python
counts = [979,120,12]
fueltype = ('Petrol','Diesel','CNG')
index = np.arange(len(fueltype))

print(counts)
print(index)
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.xticks(index , fueltype , rotation=90)
plt.show()
```

**"Using Seaborn":-**

```python
sns.set(style='darkgrid')
plt.xlabel('Age(months)')
plt.ylabel('Price(Euros)')
sns.regplot(x=a['Age'],y=a['Price'])
```

```python
sns.set(style='darkgrid')
plt.xlabel('Age(months)')
plt.ylabel('Price(Euros)')
sns.regplot(x=a['Age'],y=a['Price'],marker="*",fit_reg=False)
```

```python
sns.set(style='darkgrid')
sns.lmplot(x='Age',y='Price',data=a,fit_reg=False,hue='FuelType',legend=True,palette='Set1')
```

```python
sns.displot(a['Age'],kde=False)
sns.displot(a['Age'],kde=True)
```

```python
sns.displot(a['Age'],kde=True,bins=5)
```

```python
sns.countplot(x='FuelType',data=a)
```

```python
sns.countplot(x='FuelType',data=a,hue='Automatic')
```

**"Box & Whisker Plot":-**

```python
sns.boxplot(y=a['Price'])
```

```python
sns.boxplot(x=a['FuelType'],y= a['Price'])
```

```python
sns.boxplot(x=a['FuelType'],y= a['Price'],hue='Automatic',data=a)
```

**"Pairwise Plot":-**

```python
sns.pairplot(a,kind='scatter',hue='FuelType')
plt.show()
```
