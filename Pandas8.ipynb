**"SORTING":-**

```python
import pandas as pd
```

```python
data = pd.read_excel('Result.xlsx')
print(data)
```

```python
# Sorting a particular column:-
data.sort_values('Name of Students')
```

```python
# Sorting a particular column:-
data.sort_values('ANN')           # Ascending Order{DEFAULT}
```

```python
# Sorting a particular column:-
data.sort_values('ANN',ascending=False)    # Descending Order
```

```python
# Sorting more than one column:-
data.sort_values(['Name of Students','ANN'])  
```

```python
# Sorting more than one column:-
data.sort_values(['Name of Students','ANN'],ascending=False)
```

```python
# Sorting more than one column:-

data.sort_values(['Name of Students','ANN'],ascending=[1,0])
# Descending -> 0    , Ascending -> 1
```

```python
# Sorting more than one column:-

data.sort_values(['Name of Students','ANN'],ascending=[0,1])
# Descending -> 0    , Ascending -> 1
```

**"RANKING":-**

```python
data2 = pd.read_excel('Employee.xlsx')
print(data2)
```

```python
data2['Default_Rank'] = data2['SALARY'].rank(ascending=False)
print(data2)
```

```python
data2['Grouped_Rank'] = data2.groupby('DEPT')['SALARY'].rank()
data2.sort_values(by='DEPT')
```

```python
data2['Percentile_Rank'] = data2['SALARY'].rank(ascending=False,method='first',pct=True)
data2
```
