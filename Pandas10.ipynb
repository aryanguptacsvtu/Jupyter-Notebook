**"Unique() Function":-**

```python
import pandas as pd
```

```python
data = pd.read_csv("Toyota.csv")
data
```

```python
print(pd.unique(data['KM']))
print(pd.unique(data['HP']))
print(pd.unique(data['MetColor']))
print(pd.unique(data['Automatic']))
print(pd.unique(data['Doors']))
```

**"Drop Duplicates() Funtion":-**

```python
d = pd.read_excel('Employee.xlsx')
d
```

```python
print(d.drop_duplicates(subset=['DEPT']))
```

```python
print(d.drop_duplicates(subset=['DEPT'],inplace=True))
print(d)            # To make permanent changes,use inplace= True
```

**"Membership":-**

```python
data2 = pd.read_excel("Result.xlsx")
data2
```

```python
print(data2.query('ANN in [10,78,46,71]'))
```

```python
print(data2.query('ANN not in [10,78,46,71]'))
```

```python
print(data2.query('ANN > 50 & ML not in [53,49,42,36,69,50,63]'))
```

```python
print(data2.query('95>ANN>65'))
```

**"Descriptive Analysis:-"**
